##Aplicar filtros


###1. Apartamentos que tienen 1 dormitorio y un alquiler menor a MXN 4200.

```python
seleccion1 = dataFrame['Habitaciones'] == 1
```

```python
dataFrame[seleccion1]
```

```python
seleccion2 = dataFrame['Valor'] < 4200
dataFrame[seleccion2]
```

```python
dataFrame[seleccion2].max()
```

```python
filtro1 = (seleccion1) & (seleccion2)
dataFrame[filtro1]
```

```python
dataFrame1 = dataFrame[filtro1]
```

###2. Apartamentos que tienen al menos 2 dormitorios, un alquiler menor a MXN 10500 y una superficie mayor a 70 m².

```python
filtro2 = (dataFrame['Habitaciones']>=2) & (dataFrame['Valor']<10500) & (dataFrame['Area']>70)
dataFrame[filtro2]
```

```python
dataFrame2 = dataFrame[filtro2]
```
